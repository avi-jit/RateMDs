```python
data = pd.read_csv('processed_1.csv')[['r_id','doc_id','specialty','gender','S','P','H','K']]
data.sample(5)
```

```python
data1 = data.groupby(['doc_id','specialty','gender']).apply(lambda x: x.sum())
data1.sample(3)
```

```python
counts = dict(data.groupby(['doc_id','specialty','gender']).size())
counts
```

```python
d = []
for i,row in data1.iterrows():
    tup = row.name
    count = counts[tup]
    s = round(row['S']/count,1)
    p = round(row['P']/count,1)
    h = round(row['H']/count,1)
    k = round(row['K']/count,1)
    d.append([ str(tup[0]) , tup[1] , tup[2] , [s,p,h,k] ])
```

```python
specialties_freq = {}
for i in d:
    try:
        specialties_freq[i[1]]+= 1
    except:
        specialties_freq[i[1]] = 1
specialties_freq
```

```python
f = {}
for r in d:
    try:
        f[r[1]]+=1
    except:
        f[r[1]] = 1
        
import operator
sorted_f = sorted(f.items(), key=operator.itemgetter(1),reverse=True)

threshold = 150 #190
allowed = [] # list of allowed specialties, which have frequency above threshold
for i in sorted_f:
    if i[1] >= threshold:
        allowed.append(i[0])
    else:
        break
print(allowed)
print(len(allowed))
mapping = {}
for i,s in enumerate(allowed):
    mapping[s] = i
map_g = {'m':0,'f':1}
```

```python
for i in range(len(d)-1,-1,-1):
    if(d[i][1] not in allowed):
        try:
            del d[i]
        except:
            print(i)
len(d)
```

```python
x,y = [],[[],[],[],[]]
for r in d:
    temp = [0]*len(allowed)
    temp[ mapping[r[1]] ] = 1
    temp.append(map_g[r[2]])
    x.append(temp)
    for i in [0,1,2,3]:
        y[i].append(r[3][i])
len(x[0])
```
